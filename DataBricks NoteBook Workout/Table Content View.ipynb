```python
%py

dbutils.fs.cp(
    "https://public.tableau.com/app/sample-data/mobile_os_usage.csv",
    "dbfs:/Volumes/workspace/default/usage_metrics/mobile_os_usage.csv"
)
df = spark.read.format("csv") \
    .option("inferSchema", "true") \
    .option("header", "true") \
    .option("sep", ",") \
    .load("dbfs:/Volumes/workspace/default/usage_metrics/mobile_os_usage.csv")

display(df)
```
