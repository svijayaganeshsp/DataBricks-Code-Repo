###We are going to learn some advanced options available in the different file/table write functions (Least bother)

#### Important - There are 4 typical write modes and default is error.
1. overwrite
2. append
3. ignore
4. error

####CSV Write Advanced Feature - Very Important, Important, Not Important (just try to know once for all)
**Very Important:** path, mode, compression, sep, quote, escape, header   <br>
**Important:** nullValue,quoteAll,dateFormat, timestampFormat, ignoreLeadingWhiteSpace, ignoreTrailingWhiteSpace,lineSep  <br>
**Not Important:** escapeQuotes, charToEscapeQuoteEscaping, encoding, emptyValue

```python
struct1="cid int,name string,amt double,dop string"
df1=spark.read.csv("/Volumes/workspace/default/volumewe47_datalake/malformeddata1.txt",header=False,sep=","
                                   ,comment='#',quote="'",escape='~',schema=struct1,multiLine=True,ignoreLeadingWhiteSpace=True,ignoreTrailingWhiteSpace=True)#permit malformed records and allow me to do a RCA (Root cause analysis)
df1.printSchema()
df1.show()
```

```python
df1.write.csv(path="/Volumes/workspace/default/volumewe47_datalake/outputdata/csvout",mode="overwrite",header=True,sep="|",quote="'",escape='~',quoteAll=True)
```

####JSON Write Advanced Feature - Very Important, Important, Not Important (just try to know once for all)
**Very Important:** path, mode, compression  <br>
**Important:** dateFormat, timestampFormat, lineSep, ignoreNullFields  <br>
**Not Important:** encoding

```python
from pyspark.sql.functions import *
struct1="cid int,name int,amt double,dop date"
df1=spark.read.csv("/Volumes/workspace/default/volumewe47_datalake/malformeddata1.txt",header=False,sep=","
                                   ,comment='#',quote="'",escape='~',schema=struct1,multiLine=True,ignoreLeadingWhiteSpace=True,ignoreTrailingWhiteSpace=True,dateFormat='yyyy-dd-MM')#permit malformed records and allow me to do a RCA (Root cause analysis)
#df2=df1.withColumn("dop",to_date(date_format(col("dop"),"yyyy-dd-MM"),"yyyy-MM-dd"))
df1.printSchema()
df1.show()
```

```python
df1.write.json("/Volumes/workspace/default/volumewe47_datalake/outputdata/jsonout",mode="overwrite", compression='gzip',dateFormat='yyyy-dd-MM',encoding='UTF-8',lineSep='~',ignoreNullFields=True)
```

####Serialized Data format (orc/parquet/delta(we learn indepth in project)) Write Advanced Feature - Very Important, Important, Not Important (just try to know once for all)
**Very Important:** path, mode, partitionBy, compression  <br>

```python
df1.write.orc("/Volumes/workspace/default/volumewe47_datalake/outputdata/orcout",mode="overwrite",compression='None',partitionBy="dop")
df1.write.parquet("/Volumes/workspace/default/volumewe47_datalake/outputdata/parquetout",mode="overwrite",compression='None',partitionBy="dop")
```

```python
dforc1=spark.read.orc("/Volumes/workspace/default/volumewe47_datalake/outputdata/orcout").where("dop='2025-12-18'")
dforc1.explain()
dforc1.show()
dfparquet1=spark.read.parquet("/Volumes/workspace/default/volumewe47_datalake/outputdata/parquetout").where("dop='2025-12-18'")
dfparquet1.explain()
dfparquet1.show()
```
